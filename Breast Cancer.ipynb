```python
def PieChartplotter(df,columnname):
  values = df[columnname].value_counts()
  labels = df[columnname].unique()
  pie,ax = plt.subplots(figsize=[10,10])
  patches, texts, autotexts = ax.pie(values, labels=labels, autopct='%1.2f%%', shadow=True, pctdistance=.5,colors = ['red','#40bcde'])
  plt.legend(patches, labels, loc="best")
  plt.title(columnname,color='#e8bb48', fontsize=18)
  plt.setp(texts, fontsize=20)
  plt.setp(autotexts, size=10, color='black')
  autotexts[1].set_color('black')
  plt.axis('equal')
  plt.tight_layout()
  plt.show()

PieChartplotter(df,"diagnosis")
```

```python
num_cols1 = [col for col in df1.columns if df1[col].dtype != 'object']
num_cols1
```

```python
def plotboxplot(column_name):
    plt.boxplot(df1[column_name],vert=False)
    plt.xlabel(column_name)
    plt.title("Box Plot for outliers")
    plt.show()
```

```python
for i in range(len(num_cols1)):
    plotboxplot(num_cols1[i])
```

#### Each numerical columns have outliers, we have to remove them from the dataset for better practice

```python
df2=pd.DataFrame()
```

#### Outlier Removed

```python
for i in range(len(num_cols1)):
    mean = df1[num_cols1[i]].mean()
    std = df1[num_cols1[i]].std()
    lower_bound = mean - std * 2
    upper_bound = mean + std * 2
    df2 = df1[(df1[num_cols1[i]] >= lower_bound) & (df1[num_cols1[i]] <= upper_bound)]
```

```python
df2.head()
```

```python
import seaborn as sns
corr = df2.corr(method='pearson')

mask = np.zeros_like(corr, dtype=bool)
mask[np.triu_indices_from(mask)] = True

f, ax = plt.subplots(figsize=(15, 15))


sns.heatmap(corr, mask=mask, cmap='cool_r', vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5}, annot = True)
```
